```python
train_data1=pd.read_csv('/content/train.csv')
test_data1=pd.read_csv('/content/test.csv')
```

```python
train_data=train_data1.fillna(0)
test_data=test_data1.fillna(0)
```

```python
train_data
```

```python
test_data
```

```python
loans_one_hot_enc = pd.get_dummies(train_data)
```

```python
loans_one_hot_enc
```

```python
x = loans_one_hot_enc[["b", "c",'d','f','g']]
y = loans_one_hot_enc[["i"]]
```

```python
x_train, x_test, y_train, y_test = sklearn.model_selection.train_test_split(x, y, test_size=0.2, random_state = 124)
```

```python
from sklearn.tree import DecisionTreeClassifier
DecisionTree = DecisionTreeClassifier(criterion='entropy',random_state=2,max_depth=5)
DecisionTree.fit(x_train,y_train)
#predicted_values = DecisionTree.predict(x_test)
x = DecisionTree.score(x_test,y_test)
print("DecisionTrees's Accuracy is: ", x*100)
```

```python
from sklearn.naive_bayes import GaussianNB
NaiveBayes = GaussianNB()
NaiveBayes.fit(x_train,y_train)
predicted_values = NaiveBayes.predict(x_test)
x = metrics.accuracy_score(y_test, predicted_values)
print("Naive Bayes's Accuracy is: ", x*100)
```

```python
from sklearn.ensemble import RandomForestClassifier

RF = RandomForestClassifier(n_estimators=20, random_state=0)
RF.fit(x_train,y_train)

predicted_values = RF.predict(x_test)

x = metrics.accuracy_score(y_test, predicted_values)

print("Random Forest model Accuracy is: ", x*100)
```
